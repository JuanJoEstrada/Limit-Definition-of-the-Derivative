# Limit Definition of the Derivative

```python
import numpy as np
from math import sin, cos, log
import matplotlib.pyplot as plt
```

## Formula from the limit definition of a derivative
$\lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$

```python
def limit_derivative(f, x, h):
  """
  f: function to be differentiated 
  x: the point at which to differentiate f 
  h: distance between the points to be evaluated
  """
  # compute the derivative at x with limit definition
  return (f(x+h) - f(x)) / h
```

## Define three functions
* `f1()` — which defines *f₁(x) = sin(x)*
* `f2()` — which defines *f₂(x) = x⁴*
* `f3()` — which defines *f₃(x) = x² log(x)*

The default base for the math.log() function is e, so the function defined by f3(x) is evaluated as: *f₃(x) = x² ln(x)*


```python
def f1(x):
  return sin(x)

def f2(x):
  return pow(x, 4)

def f3(x):
  return pow(x, 2) * log(x)
```

## We are going to graph approximations using **plot_approx_deriv()**

```python
def plot_approx_deriv(f):
  x_vals = np.linspace(1, 10, 200)
  h_vals = [10, 5, 1, .25]
  true_derivative = None

  for h in h_vals:
      derivative_values = []
      for x in x_vals:
          derivative_values.append(limit_derivative(f, x, h))
      plt.plot(x_vals, derivative_values, linestyle='--', label="h=" + str(h) )

  if f == f1:
    # f1 = sin(x)
    true_derivative = [cos(val) for val in x_vals]
  elif f == f2:
    # f2 = x^4
    true_derivative = [4 * pow(val, 3) for val in x_vals]
  else:
    # f3 = x^2*log(x)
    true_derivative = [pow(val, 2)*(1/val) + 2*pow(val,1)*log(val) for val in x_vals]
    
  plt.plot(x_vals, true_derivative, label="True Derivative", linewidth=4, alpha=0.5)
  plt.legend()
  plt.title("Convergence to Derivative by varying h")
  plt.show()
  plt.clf()

plot_approx_deriv(f1)
plot_approx_deriv(f2)
plot_approx_deriv(f3)
```

## Conclusion
As the value of h gets closer to zero, the limit derivative of f1 approaches the true derivative. This aligns with the limit definition of a derivative.
